```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df = pd.read_csv('cleaned_hotel_bookings.csv')
print(f"Successfully loaded {len(df)} booking records.\n")
```

# **TEST 1: TWO-SAMPLE T-TEST (Lead Time vs Attrition)**

```python
# Grouping lead times by booking status
canceled_lead = df[df['is_canceled'] == 1]['lead_time']
checked_in_lead = df[df['is_canceled'] == 0]['lead_time']
```

```python
# Run Welch's T-Test (equal_var=False handles unequal group sizes)
t_stat, p_val_t = stats.ttest_ind(canceled_lead, checked_in_lead, equal_var=False)

print(f"Average Lead Time for Canceled Bookings: {canceled_lead.mean():.2f} days")
print(f"Average Lead Time for Checked-In Bookings: {checked_in_lead.mean():.2f} days")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_val_t}")
```

```python
# Interpretation
alpha = 0.05
if p_val_t < alpha:
    print("Conclusion: Reject Null Hypothesis (H0). Lead time significantly impacts cancellations.")
else:
    print("Conclusion: Fail to Reject Null Hypothesis (H0). No significant difference found.")
```

# TEST 2: CHI-SQUARE TEST (Loyalty vs Attrition)

```python
# Create a contingency table (Cross-tabulation)
contingency_table = pd.crosstab(df['is_repeated_guest'], df['is_canceled'])
print("Contingency Table (Observed Frequencies):")
print(contingency_table)
```

```python
# Run the Chi-Square Test of Independence
chi2_stat, p_val_chi2, dof, expected = stats.chi2_contingency(contingency_table)

print(f"\nChi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_val_chi2}")
```

```python
# Interpretation
if p_val_chi2 < alpha:
    print("Conclusion: Reject Null Hypothesis (H0). Guest loyalty and cancellation behaviors are strongly dependent.")
else:
    print("Conclusion: Fail to Reject Null Hypothesis (H0). No significant relationship found.")
```
